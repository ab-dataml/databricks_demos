### set latest model to champion 

```python
from mlflow.tracking import MlflowClient

client = MlflowClient()

# Unity Catalog uses search_model_versions instead
versions = client.search_model_versions(
    filter_string="name='fraud_demo.transactions.fraud_classifier'"
)

latest = sorted(versions, key=lambda v: v.creation_timestamp, reverse=True)[0]

client.set_registered_model_alias(
    name="fraud_demo.transactions.fraud_classifier",
    alias="Champion",
    version=latest.version
)

print(f"Champion → version {latest.version}")
```
