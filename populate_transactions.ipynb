```python
import random, time, uuid
from pyspark.sql import Row
from datetime import datetime

def generate_txn():
    is_fraud = random.random() < 0.02  # 2% fraud rate
    return Row(
        txn_id=str(uuid.uuid4()),
        user_id=f"u{random.randint(1,500)}",
        amount=round(random.uniform(500,5000) if is_fraud
                     else random.uniform(1,200), 2),
        merchant=random.choice(["Amazon","Shell","Walmart","Unknown"]),
        hour=datetime.now().hour,
        is_fraud=int(is_fraud),
        ts=datetime.now().isoformat()
    )

rows = [generate_txn() for _ in range(10_000)]
df = spark.createDataFrame(rows)
df.write.format("delta").mode("overwrite") \
  .saveAsTable("fraud_demo.transactions.raw")
```
